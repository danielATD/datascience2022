```python
#Estilo matlab

x = np.linspace(0,10,100)
plt.figure()

plt.subplot(2,2,1) #(rows, columns, numero_De_panel)
plt.plot(x, np.sin(x))

plt.subplot(2,2,2)
plt.plot(x,np.cos(x))

plt.subplot(2,2,3)
plt.plot(x,np.cos(x))
```

```python
#Estilo orientado a objetos
fig,ax = plt.subplots(2,1) #(rows, columns)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```
